### 시각화 패키지 : seaborn 사용 

```python
%matplotlib inline
import seaborn as sns
import pandas as pd
```

### load data

```python
train=pd.read_csv("/Users/yuhyun/Downloads/train.csv")
```

### countplot 

```python
sns.countplot(data=train, x="Embarked", hue="Survived") #hue option은 종류가 3개가 넘어가면 의미가 없어진다. 
```

```python
pd.pivot_table(train, index="Pclass", values="Fare") #embarked 에 따른 Fare의 값
```

```python
sns.countplot(data=train,x="Pclass", hue="Survived")
```

### barplot ; x 축이 서로 연결선이 없을때 사용

```python
sns.barplot(data=train, x="Pclass",y="Fare",hue="Survived")
```

### point plot


```python
sns.pointplot(data=train,x="Pclass" ,y="Fare",hue="Survived")
```

### point plot : x=시간 y=매출 같은 예제처럼 
### 시간의 변화별 매출의 변화를 알수있는 데이터 

### lmplot

```python
sns.lmplot(data=train, x="Age",y="Fare",hue="Survived", fit_reg=False)
```

```python
train["Fare"].mean()
```

```python
low_fare=train[train["Fare"]<100] #ouliter 지우기 
```

```python
sns.lmplot(data=low_fare,x="Age", y="Fare",
hue="Survived", fit_reg=False)
```

### distplot 
 

```python
sns.distplot(train["Fare"])
```

```python
sns.distplot(low_fare["Fare"])

```

```python
sns.boxplot(x="Pclass", y="Fare",
            hue="Survived", palette=["m", "g"],
            data=train)
```
